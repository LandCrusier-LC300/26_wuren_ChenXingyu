```python
'''
模型训练基本流程：第1步：导入需要的库
   ↓
第2步：加载数据
   ↓
第3步：数据预处理（切分训练集/测试集）
   ↓
第4步：创建并训练模型
   ↓
第5步：预测并评估结果
'''
from sklearn.datasets import fetch_california_housing   #从sklearn库导入加州房价数据集
from sklearn.model_selection import train_test_split    #导入用与划分数据集的模块（将数据集分为训练集合和测试集）
from sklearn.linear_model import LinearRegression  #导入线性回归模型
from sklearn.metrics import mean_squared_error,r2_score

#加载数据
housing=fetch_california_housing(as_frame=True)
x=housing.data
y=housing.target

#划分数据集
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
#将数据集按5：1的比例划分为训练集和测试集

#训练模型
model=LinearRegression()
model.fit(x_train,y_train)

#预测和评估
y_pred = model.predict(x_test)
print("均方误差{0}",format(mean_squared_error(y_test,y_pred)))
print("分数{0}",format(r2_score(y_test,y_pred)))
```

```python
'''
模型训练基本流程：第1步：导入需要的库
   ↓
第2步：加载数据
   ↓
第3步：数据预处理（切分训练集/测试集）
   ↓
第4步：创建并训练模型
   ↓
第5步：预测并评估结果
'''
from sklearn.datasets import fetch_california_housing   #从sklearn库导入加州房价数据集
from sklearn.model_selection import train_test_split    #导入用与划分数据集的模块（将数据集分为训练集合和测试集）
from sklearn.linear_model import LinearRegression,Ridge #导入线性回归模型，和正则化惩罚模块Ridge
from sklearn.metrics import mean_squared_error,r2_score

#加载数据
housing=fetch_california_housing(as_frame=True)
x=housing.data
y=housing.target

#划分数据集
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
#将数据集按5：1的比例划分为训练集和测试集

#训练模型
model=Ridge(alpha=1.0)#alpha越大，代表惩罚力度越大
model.fit(x_train,y_train)

#预测和评估
y_pred = model.predict(x_test)
print("均方误差{0}",format(mean_squared_error(y_test,y_pred)))
print("分数{0}",format(r2_score(y_test,y_pred)))
```

```python
'''
模型训练基本流程：第1步：导入需要的库
   ↓
第2步：加载数据
   ↓
第3步：数据预处理（切分训练集/测试集）
   ↓
第4步：创建并训练模型
   ↓
第5步：预测并评估结果
'''
from sklearn.datasets import fetch_california_housing   #从sklearn库导入加州房价数据集
from sklearn.model_selection import train_test_split    #导入用与划分数据集的模块（将数据集分为训练集合和测试集）
from sklearn.linear_model import LinearRegression,Ridge #导入线性回归模型，和正则化惩罚模块Ridge
from sklearn.metrics import mean_squared_error,r2_score

#加载数据
housing=fetch_california_housing(as_frame=True)
x=housing.data
y=housing.target

#划分数据集
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
#将数据集按5：1的比例划分为训练集和测试集

#训练模型
model=Ridge(alpha=1.0)#alpha越大，代表惩罚力度越大
model.fit(x_train,y_train)

#预测和评估
y_pred = model.predict(x_test)
print(f"均方误差: {mean_squared_error(y_test, y_pred):.4f}")
print(f"R² 分数: {r2_score(y_test, y_pred):.4f}")
```
