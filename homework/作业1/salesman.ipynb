```python
# -*- coding: utf-8 -*-
from __future__ import print_function
import math
import random
from collections import defaultdict
from simanneal import Annealer  #模拟退火库

#函数：实现两个经纬坐标之间距离的计算
'''输入值为地球上两个地方的经纬坐标，通过math.radians方法将度数值转化为弧度值，然后通过球面余弦定律计算两个地方之间在地球上的距离
球面余弦定律：将经纬坐标转化为空间直角坐标：x=cos(lat)cos(lon),y=cos(lat)sin(lon),z=sin(lat)sin(lon)。从而分别得到两个地方的直角坐标，即对应的向量r1和r2，
两个向量为单位向量，数量积为两个向量夹角的余弦值。从而得到cos(夹角)=cos(lat1)cos(lat2)*cos(lon1-lon2)+sin(lat1)sin(lat2)。该夹角与半径乘积即为两地点的圆弧距离
'''
def distance(a, b):  #输入值为两个地方的经纬坐标
    """Calculates distance between two latitude-longitude coordinates."""
    R = 3963  # radius of Earth (miles)---地球半径（米）
    #将经纬度数转化为弧度
    lat1, lon1 = math.radians(a[0]), math.radians(a[1])
    lat2, lon2 = math.radians(b[0]), math.radians(b[1])
    #通过几何关系计算两地之间的距离：
    return math.acos(math.sin(lat1) * math.sin(lat2) +
                     math.cos(lat1) * math.cos(lat2) * math.cos(lon1 - lon2)) * R

#定义旅行商问题：假设要去若干个城市（城市之间的距离已知），该如何规划路线来确保每个城市只去一次的同时缩短时间和距离
class TravellingSalesmanProblem(Annealer):  #定义旅行商类，继承Annealer类

    """Test annealer with a travelling salesman problem.
    """

    # pass extra data (the distance matrix) into the constructor
    def __init__(self, state, distance_matrix):
        self.distance_matrix = distance_matrix
        super(TravellingSalesmanProblem, self).__init__(state)  # important!  #继承所导入的Annealer里面的state
    #功能：先获得总路线的长度，再随机找两个城市，从一个城市走到另一个城市，返回新路线路径与原来路径的长度差值
    def move(self):
        """Swaps two cities in the route."""
        # no efficiency gain, just proof of concept
        # demonstrates returning the delta energy (optional)
        initial_energy = self.energy()
        #随机找两个不同的城市并且在原有路径中调换他们的位置，这样就获得了一个新的路径
        a = random.randint(0, len(self.state) - 1)
        b = random.randint(0, len(self.state) - 1)
        self.state[a], self.state[b] = self.state[b], self.state[a]

        return self.energy() - initial_energy
    #计算总路线的长度
    def energy(self):
        """Calculates the length of the route."""
        e = 0
        for i in range(len(self.state)):  #遍历所有城市
            e += self.distance_matrix[self.state[i-1]][self.state[i]]  #从距离矩阵中找出当前城市和上一个城市的距离，然后累加到e
        return e


if __name__ == '__main__':

    # latitude and longitude for the twenty largest U.S. cities  #列举城市以及城市的经纬坐标
    cities = {
        'New York City': (40.72, 74.00),
        'Los Angeles': (34.05, 118.25),
        'Chicago': (41.88, 87.63),
        'Houston': (29.77, 95.38),
        'Phoenix': (33.45, 112.07),
        'Philadelphia': (39.95, 75.17),
        'San Antonio': (29.53, 98.47),
        'Dallas': (32.78, 96.80),
        'San Diego': (32.78, 117.15),
        'San Jose': (37.30, 121.87),
        'Detroit': (42.33, 83.05),
        'San Francisco': (37.78, 122.42),
        'Jacksonville': (30.32, 81.70),
        'Indianapolis': (39.78, 86.15),
        'Austin': (30.27, 97.77),
        'Columbus': (39.98, 82.98),
        'Fort Worth': (32.75, 97.33),
        'Charlotte': (35.23, 80.85),
        'Memphis': (35.12, 89.97),
        'Baltimore': (39.28, 76.62)
    }

    # initial state, a randomly-ordered itinerary
    init_state = list(cities)  #把前面字典的建（城市）转化为列表
    random.shuffle(init_state)   #把城市列表打乱

    # create a distance matrix    #生成距离矩阵，计算两个城市之间的距离，放到矩阵中对应的位置中
    distance_matrix = defaultdict(dict)
    for ka, va in cities.items():
        for kb, vb in cities.items():
            distance_matrix[ka][kb] = 0.0 if kb == ka else distance(va, vb)  #如果是同一个城市，则值为0,否则为两个城市之间的距离

    tsp = TravellingSalesmanProblem(init_state, distance_matrix)#实例化旅行商类
    tsp.set_schedule(tsp.auto(minutes=0.2))#设置参数，让程序自动运行0.2秒
    # since our state is just a list, slice is the fastest way to copy
    tsp.copy_strategy = "slice"
    state, e = tsp.anneal()   #返回最有路径和他的总距离
    #固定第一个城市为纽约
    while state[0] != 'New York City':  #如果第一个城市不是纽约
        state = state[1:] + state[:1]  # rotate NYC to start，就把第一个城市放在列表最后
    #打印最优路线的总距离及具体路线
    print()
    print("%i mile route:" % e)
    print(" ➞  ".join(state))
```
