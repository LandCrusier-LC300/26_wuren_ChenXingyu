```python
'''
SVM分类基本流程：第1步：导入需要的库
   ↓
第2步：加载数据
   ↓
第3步：数据预处理（切分训练集/测试集）
   ↓
第4步：创建并训练模型
   ↓
第5步：预测并评估结果
'''

from sklearn.datasets import load_iris           # 从sklearn库导入鸢尾花数据集
from sklearn.model_selection import train_test_split    # 导入用于划分数据集的模块
from sklearn.svm import SVC                     # 导入SVM分类模型
from sklearn.metrics import accuracy_score      # 导入准确率评估指标

# 加载数据
iris = load_iris()
x = iris.data
y = iris.target

# 划分数据集
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# 训练模型
model = SVC()
model.fit(x_train, y_train)

# 预测和评估
y_pred = model.predict(x_test)
print(f"准确率: {accuracy_score(y_test, y_pred):.4f}") 
```
